```python
from icrawler.builtin import GoogleImageCrawler
import os

def download_images(keyword, folder, max_num = 150):
    os.makedirs(folder, exist_ok=True)
    crawler = GoogleImageCrawler(storage={'root_dir':folder})
    crawler.crawl(keyword=keyword, max_num=max_num)

download_images('wild bird in nature photography','images/bird', max_num=20)
download_images('forest landscape trees photography','images/forest', max_num=20)

```
